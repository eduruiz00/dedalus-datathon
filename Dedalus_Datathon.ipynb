```python
import os
import numpy as np
import pandas as pd
import sklearn.preprocessing
import matplotlib.pyplot as plt
import math
from sklearn.metrics import r2_score, classification_report, confusion_matrix, accuracy_score
import seaborn as sns
from sklearn.model_selection import GridSearchCV
from imblearn.ensemble import RUSBoostClassifier
import warnings
warnings.filterwarnings('ignore')
```

```python
path="data"
tables_name=os.listdir(path)
print(tables_name)
```

```python
tables={}
for element in tables_name:
    tables[element[0:element.find(".csv")]]=pd.read_csv(path+"/"+element)
```

```python
table_name=list(tables.keys())
```

```python
def le(col):
    col_list=list(col)
    labels={}
    for i in range(len(col_list)):
        if col_list[i] in labels:
            next
        else:
            labels[col_list[i]]=len(labels)
    col=col.replace(labels)
    return col, labels
```

```python
pat_ex=1335511
for element in table_name:
    try:
        tabla=tables[element]
        print("**{}**".format(element))
        display(tabla[tabla["patientunitstayid"]==pat_ex])
    except:
        next
```

```python
def generate_data(tables):
    
    input_information = pd.merge(tables["patient"], tables["hospital"], on="hospitalid", how="left")
    input_information = input_information.drop('hospitalid', axis=1)
    
    inter_col = [0,2,3,4,6,7,10,16,18,21,28,29,30]
    dic_dics = {}
    ii1 = input_information.iloc[:,inter_col]
    col_types = ii1.dtypes.values
    for i in range(0,len(col_types)):
        tipe = col_types[i]
        if tipe=="O" and i!=2:
            ii1.iloc[:,i],dic_dics[ii1.iloc[:,i].name] = le(ii1.iloc[:,i])
    ii1["age"][ii1["age"]=="> 89"]=90
    
    usids = []
    allergies = []
    adm_drugs = []
    micro_lab= []
    lab= []
    pastHistory = []
    for i in range(0,len(input_information)):
        usid = input_information["patientunitstayid"][i]
        if len(tables["allergy"][tables["allergy"]["patientunitstayid"]==usid])>0:
            allergies.append(1)
        else:
            allergies.append(0)
        usids.append(usid)
        if len(tables["admissiondrug"][tables["admissiondrug"]["patientunitstayid"]==usid])>0:
            adm_drugs.append(len(np.unique(tables["admissiondrug"]["drughiclseqno"][tables["admissiondrug"]["patientunitstayid"]==usid])))
        else:
            adm_drugs.append(0)
        if len(tables["lab"][tables["lab"]["patientunitstayid"]==usid])>0:
            lab.append(1)
        else:
            lab.append(0)
        if len(tables["microLab"][tables["microLab"]["patientunitstayid"]==usid])>0:
            micro_lab.append(1)
        else:
            micro_lab.append(0)

    pres_allergy = pd.DataFrame()
    pres_allergy["patientunitstayid"] = usids
    pres_allergy["PresenceAllergy"] = allergies
    pres_adm_drug=pd.DataFrame()
    pres_adm_drug["patientunitstayid"] = usids
    pres_adm_drug["PresenceAdmDrug"] = adm_drugs
    pres_lab = pd.DataFrame()
    pres_lab["patientunitstayid"] = usids
    pres_lab["PresenceLab"] = lab
    pres_microlab = pd.DataFrame()
    pres_microlab["patientunitstayid"] = usids
    pres_microlab["PresenceMicrolab"] = micro_lab
    ii1 = pd.merge(ii1, pres_allergy, on="patientunitstayid", how="left")
    ii1 = pd.merge(ii1, pres_adm_drug, on="patientunitstayid", how="left")
    ii1 = pd.merge(ii1, pres_lab, on="patientunitstayid", how="left")
    ii1 = pd.merge(ii1, pres_microlab, on="patientunitstayid", how="left")
    
    apach = tables["apacheApsVar"][1:]
    ii1 = pd.merge(ii1, apach, on="patientunitstayid", how="left")

    
    usids=[]
    n_days=[]
    Apchr=tables["apachePatientResult"]
    cont=0
    for i in range(len(ii1)):
        id_ = ii1["patientunitstayid"][i]
        try:
            n_days.append(Apchr[Apchr["patientunitstayid"] == id_]["actualhospitallos"].iloc[0])
        except:
            n_days.append(np.nan)
        usids.append(id_)
    y_days=pd.DataFrame()
    y_days["patientunitstayid"]=usids
    y_days["UCIdays"]=n_days
    ii1 = pd.merge(ii1, y_days, on="patientunitstayid", how="left")
    
    treat=[]
    tr=tables["treatment"]
    for i in range(0,len(ii1)):
        usid=ii1["patientunitstayid"][i]
        if len(tr[tr["patientunitstayid"] == usid])>0:
            treat.append(len(tr[tr["patientunitstayid"] == usid]))
        else:
            treat.append(0)
    pres_treat=pd.DataFrame()
    pres_treat["patientunitstayid"]=usids
    pres_treat["treatment"]=treat
    ii1=pd.merge(ii1, pres_treat, on="patientunitstayid", how="left")
    
    past=[]
    tr=tables["pastHistory"]
    for i in range(0,len(ii1)):
        usid=ii1["patientunitstayid"][i]
        if len(tr[tr["patientunitstayid"] == usid])>0:
            past.append(len(tr[tr["patientunitstayid"] == usid]))
        else:
            past.append(0)
    past_treat=pd.DataFrame()
    past_treat["patientunitstayid"]=usids
    past_treat["pastHistory"]=past
    ii1=pd.merge(ii1, past_treat, on="patientunitstayid", how="left")

    
    return ii1
```

# N days ICU

```python
ii1=generate_data(tables)
ii1 = ii1[ii1['UCIdays'].notna()]
ii1 = ii1.fillna(-1)
```

```python
ii1 = ii1[ii1['UCIdays'].notna()]
ii1["days_ranges"] = pd.cut(ii1["UCIdays"],bins=[0,5,100],right=False)
ii1["days_ranges"].value_counts()
```

```python
from scipy.stats import gmean
plt.figure(dpi=150)
plt.title("Stay in the ICU")
plt.hist(ii1['UCIdays'],bins=50,color="#007AA8")
plt.ylabel("Count")
plt.xlabel("Time (Days)")
plt.axvline(gmean(ii1['UCIdays']),color="#301934")
plt.show()
print("GMEAN: ",gmean(ii1['UCIdays']))
```

```python
from sklearn.model_selection import train_test_split
x=ii1.drop(["UCIdays",'days_ranges',"treatment",'PresenceLab','PresenceMicrolab'],axis=1)
y,dic_out=le(ii1["days_ranges"])
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=1,test_size=0.10)
```

```python
'''
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
param_grid = {'n_estimators': [1,50, 100, 200,500],'max_depth': [10, 50, 100],'bootstrap': [True, False]}
grid = GridSearchCV(RandomForestClassifier(random_state=1),param_grid,refit=True,verbose=2)
grid.fit(X_train,y_train)
print(grid.best_estimator_)
'''
```

```python
from sklearn.ensemble import RandomForestClassifier
r_def_1 = RandomForestClassifier(bootstrap=False, max_depth=50, n_estimators=200,random_state=1)
r_def_1.fit(X_train, y_train)
Y_pred = r_def_1.predict(X_test)
print(classification_report(y_test, Y_pred))
cm = confusion_matrix(y_test, Y_pred)
ax= plt.subplot()
sns.heatmap(cm, annot=True, ax = ax,cmap="viridis", fmt='.4g' )
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels') 
ax.set_title('Confusion Matrix for ICU LOS')
ax.xaxis.set_ticklabels(["- 5 days","+ 5 days"]); ax.yaxis.set_ticklabels(["- 5 days","+ 5 days"])
plt.show()
print(accuracy_score(y_test, Y_pred)*100)
```

```python
from sklearn import metrics
metrics.RocCurveDisplay.from_estimator(r_def_1, X_test, y_test) 
plt.title("ROC curve")
plt.show()
```

```python
features = X_train.columns
importances = r_def_1.feature_importances_ # Impurity-based feature importances
indices = np.argsort(importances) # Put feature with biggest importance at the top
plt.title('Feature Importances')
plt.barh(range(len(indices))[8:16], importances[indices][8:16], color='#000080')
plt.yticks(range(len(indices))[8:16], [features[i] for i in indices][8:16])
plt.xlabel('Relative Importance')
plt.show()
```

# Multiclass problem

```python
ii1=generate_data(tables)
ii1 = ii1.fillna(-1)
clases=[]
clases=list(tables["treatment"]["treatmentstring"].value_counts().keys()[0:20])
clases.append("other")
clases.append("No treatment")
pe , dict_treats = le(pd.Series(clases))
treatments_2=tables["treatment"]
treat_output = np.zeros((len(ii1),22))
for i in range(len(ii1)):
    usid = ii1["patientunitstayid"][i]
    treat_pat = treatments_2[tables["treatment"]["patientunitstayid"] == usid]
    if len(treat_pat) > 0:
        treatments = treat_pat["treatmentstring"]
        for m in range(len(treatments)):
            if treatments.iloc[m] in dict_treats:
                treat_output[i,dict_treats[treatments.iloc[m]]]=1
            else:
                treat_output[i,20]=1
    else:
        treat_output[i,21]=1
x = ii1.drop(["UCIdays","treatment",'PresenceLab','PresenceMicrolab'],axis=1)
y = treat_output
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=1,test_size=0.10)
```

```python
from sklearn.ensemble import RandomForestClassifier
r_multi = RandomForestClassifier(bootstrap=True, max_depth=50, n_estimators=350,random_state=1)
r_multi.fit(X_train, y_train)
predictions=r_multi.predict_proba(X_test)
predicted=np.zeros((y_test.shape))
for i in range(len(predictions)):
    if predictions[i].shape[1]==2:
        predict=np.round(predictions[i],2)[:,1]
        predicted[:,i]=predict
    else:
        predict=np.round(predictions[i],2)
        predicted[:,i]=predict[:,0]
```

```python
for i in range(100):
    threshold=0+0.01*i
    predict_2=predicted
    predict_2[predict_2>threshold]=1
    predict_2[predict_2<threshold]=0
    print(np.sum(predict_2 == y_test)/(predicted.shape[0]*predicted.shape[1]))
```
