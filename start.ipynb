```python
import sqlite3
con = sqlite3.connect('data-mysqlite/eicu_v2_0_1.sqlite3')

cur = con.cursor()
```

```python
patients = cur.execute('''SELECT * FROM patient''')
patients_cols = list(map(lambda x: x[0], cur.description))
patients = pd.DataFrame(patients, columns=patients_cols)
patients
```

```python
patients_int = ['patientunitstayid', 'gender', 'age', 'ethnicity', 'hospitalid', 'admissionheight', 'hospitaladmitsource', 'hospitaldischargelocation', 'unittype', 'unitadmitsource','admissionweight']
input_information = patients[patients_int]
input_information
```

```python
hospitals = cur.execute('''SELECT * FROM hospital''')
hospitals_cols = list(map(lambda x: x[0], cur.description))
hospitals = pd.DataFrame(hospitals, columns=hospitals_cols)
hospitals
```

```python
input_information = pd.merge(input_information, hospitals, on="hospitalid", how="left")
input_information
```
