```python
import sqlite3
import gzip

'''
sqlite_doc = gzip.open('data-mysqlite/eicu_v2_0_1.sqlite3.gz', 'rb')
file_content = sqlite_doc.read()
'''
data = gzip.GzipFile('data-mysqlite/eicu_v2_0_1.sqlite3.gz','r')
d = data.read()

con = sqlite3.connect(d)

cur = con.cursor()
```

```python
patients = cur.execute('''SELECT * FROM patient''')
patients_cols = list(map(lambda x: x[0], cur.description))
patients = pd.DataFrame(patients, columns=patients_cols)
patients
```

```python
patients_int = ['patientunitstayid', 'gender', 'age', 'ethnicity', 'hospitalid', 'admissionheight', 'hospitaladmitsource', 'hospitaldischargelocation', 'unittype', 'unitadmitsource','admissionweight']
input_information = patients[patients_int]
input_information
```

```python
hospitals = cur.execute('''SELECT * FROM hospital''')
hospitals_cols = list(map(lambda x: x[0], cur.description))
hospitals = pd.DataFrame(hospitals, columns=hospitals_cols)
hospitals
```

```python
input_information = pd.merge(input_information, hospitals, on="hospitalid", how="left")
input_information = input_information.drop('hospitalid', axis=1)
input_information
```

```python
allergies = cur.execute('''SELECT * FROM allergy''')
allergies_cols = list(map(lambda x: x[0], cur.description))
allergies = pd.DataFrame(allergies, columns=allergies_cols)
allergies
```

```python
allergies['allergyname'].unique()
```

```python
allergies['num'] = 1
patients_allergies = allergies.pivot_table(index=['patientunitstayid'], 
            columns=['allergyname'], values='num').fillna(0)
patients_allergies
```
